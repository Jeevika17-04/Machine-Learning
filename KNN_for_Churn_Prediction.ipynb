```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML data set/Copy of WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
df.isna().sum()
```

```python
df.nunique()
```

```python
df.drop(axis = 1, columns = ['customerID'], inplace = True)
```

```python
monthlyCharges = df['MonthlyCharges']
tenure = df['tenure']
df.drop(axis = 1, columns = ['MonthlyCharges', 'tenure'], inplace = True)
```

```python
cols = list(df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in cols:
  df[col] = le.fit_transform(df[col])

df
```

```python
y = df.iloc[:, -1]
df.drop(axis = 1, columns = ['Churn'], inplace = True)
df['MonthlyCharges'] = monthlyCharges
df['tenure'] = tenure
x = df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print(accuracy_score(y_test, y_pred))
```
